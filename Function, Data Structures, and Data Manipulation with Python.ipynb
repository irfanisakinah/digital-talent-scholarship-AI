```python
#Function in python
def f(x): 
    return x**2 
f(8)
```

```python
#lambda function
g = lambda x: x**2 
g(8)
```

```python
#defining even function and applying list and filter function
def even(x):     
    if x % 2 != 0:
        return True
    else: 
        return False 
list(filter(even, range(0,30))) 
```

```python
#defining square function
def square(x): 
    return x**2 

list(map(square, range(0,11))) 
```

```python
#defining expo function
def expo(x, y): 
    return x**y 
list(map(expo, range(0,5), range(0,5))) 
```
