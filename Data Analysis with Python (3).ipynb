```python
import pandas as pd
df = pd.read_csv('imports-85.data', header=None)
df.info()
```

```python
#Exercise 1
#Handling incorrect type assignment
import numpy as np
for i in range(len(df.columns)):
    if (i!=5) and (df[i].dtypes=='object') and (df[df[i]=='?'].shape[0]>0):
    #not including att-5 (num-of-doors)
        print('\nAttribute-',i,':',df[df[i]=='?'].shape[0])
        df[i] = df[i].replace('?', np.NaN)
        df[i] = df[i].astype('float64')
df.dtypes
```

```python
#Exercise 2
#Drop missing values
df_drop = df.dropna(axis='columns') #Drop the columns where at least 1 element is missing.
df_drop.shape
```

```python
df_drop = df.dropna(how='all') #Drop the rows where all elements are missing.
df_drop.shape
```

```python
df_drop = df.dropna(thresh=24) # Keep only the rows with at least 24 non-NA values.
df_drop.shape
```

```python
#Exercise 3
#Replace missing values
for i in range(len(df.columns)):
    if (df[df[i].isnull()].shape[0]>0):
        print('\nAttribute-',i,' (before) :',df[df[i].isnull()].shape[0])
        if (df[i].dtypes in ['int64','float64']):
            df[i].fillna(df[df[i].notnull()][i].mean(), inplace=True)
        else:
            df[i].fillna(df[df[i].notnull()][i].mode(), inplace=True)
        print('\nAttribute-',i,' (after) :',df[df[i].isnull()].shape[0])
df.describe()
```

```python
#Exercise 4
#Add new features
df['fueleconomy'] = 0.55*df[23]+0.45*df[24]
df['fueleconomy']
df
#new attribute describes the combination of values between city and highway mpg
```

```python
#MinMaxScaler
from sklearn.preprocessing import MinMaxScaler
min_max_scaler = MinMaxScaler()
newdf = df.select_dtypes(include=['float64'])
newdf_scaled = min_max_scaler.fit_transform(newdf)
newdf =pd.DataFrame(newdf_scaled)
newdf.head()
```

```python
#StandardScaler
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
df_std = df.select_dtypes(include=['float64'])
df_std = df_std.dropna()
df_std=sc.fit_transform(df_std)
df_std = pd.DataFrame(df_std)
df_std.head()
```

```python
#Label Encoding
df = pd.read_csv('imports-85.data', header=None)
for i in range(len(df.columns)):
    if (df[i].dtypes=='object'):
        df[i] = df[i].astype('category')
        df[i] = df[i].cat.codes
df.head()
```

```python
#One-hot encoding
df_encode = pd.get_dummies(data=df, columns=[2])
df_encode.head()
```
